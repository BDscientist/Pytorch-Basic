# 8. Neural Networks

```python
import numpy as np

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.init as init

from torch.autograd import Variable
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## 8.1 Generate Data

```python
x = torch.linspace(-1, 1, 500)
y = x.pow(3) + 0.2*torch.rand(x.size())
x, y = Variable(x), Variable(y)

plt.scatter(x.data.numpy(), y.data.numpy())
plt.show()
```

```python
x.shape, y.shape
```

```python
x = torch.unsqueeze(x, dim=1)
y = torch.unsqueeze(y, dim=1)
```

## 8.2 Define Model

```python
hidden1 = nn.Linear(1, 10, bias=True)
activation1 = nn.ReLU()

hidden2 = nn.Linear(10, 20, bias=True)
activation2 = nn.ReLU()

hidden3 = nn.Linear(20, 10, bias=True)
activation3 = nn.ReLU()

output = nn.Linear(10, 1, bias=True)
```

```python
model = nn.Sequential(hidden1, activation1, hidden2, activation2, hidden3, activation3, output)
```

```python
loss = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.1)
```

## 8.3 Train Model

```python
for t in range(2000):
    pre = model(x)
    cost = loss(pre, y)
    
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()
    
    if t % 400  == 0:
        plt.scatter(x.data.numpy(), y.data.numpy())
        plt.plot(x.data.numpy(), pre.data.numpy(), 'r--', lw=4)
        plt.text(0.5, 0, 'Loss=%.4f' % cost.data[0], fontdict={'size':10, 'color':'red'})
        plt.show()
```
